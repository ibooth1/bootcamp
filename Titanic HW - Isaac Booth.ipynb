### load the titanic dataset & convert the gender column into a numeric field

```python
import pandas as pd

loc = "fallcampdata/titanic3.xls"
df = pd.read_excel(loc)

def sexnum(x):
    if x == "female":
        return 0
    if x == "male":
        return 1

df["sex"] = df["sex"].apply(sexnum)
df.head()
```

### create a column that gives total number of relatives on board

```python
def totrel(x,y):
    return x+y

df["total relatives"] = totrel(df["sibsp"],df["parch"])
df.head
```
